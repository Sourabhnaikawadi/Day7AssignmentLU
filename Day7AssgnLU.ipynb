Q1. Use of dictionary.and convert values to keys.

```
port1 = {21:"FTP", 22:"SSH",23:"telnet",80:"http"}
new_port1= {value : key for (key, value)in port1.items()}
print(new_port1)
```

Q2. Take a tuple and make new list which contains sum of elements in tuple.

```
tup1 = [(1,2),(3,4),(5,6),(4,5)]
res = list(map(sum, list(tup1)))
print(res)
```

Q3 A list contains tuple and lists. Make elements to the inner lists and tuple to outer lists.

```
list1 = [(1,2,3),[1,2],['a','hit','less']]
flatlist = []
for elem in list1:
    flatlist.extend(elem)
print('Flat List : ', flatlist)
```
